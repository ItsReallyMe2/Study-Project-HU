## IMPORT ALL NECESSARY DEPENDENCIES

```python
from sklearn.datasets import load_iris, load_diabetes, load_wine
import custom_func as pc
from sklearn.decomposition import PCA
```

## LOAD THE DATASETS FROM SCIKIT LEARN AS DATAFRAME

```python
iris_data, iris_target = load_iris(return_X_y=True, as_frame=True)
#JOIN TARGET AND SET IT AS INDEX FOR VISUALISATION 
iris_data_target = iris_data.join(iris_target).set_index('target')

diabetes_data, diabetes_target = load_diabetes(return_X_y=True, as_frame=True)
#JOIN TARGET AND SET IT AS INDEX FOR VISUALISATION 
diabetes_data_target = diabetes_data.join(diabetes_target).set_index('target')

wine_data, wine_target = load_wine(return_X_y=True, as_frame=True)
#JOIN TARGET AND SET IT AS INDEX FOR VISUALISATION 
wine_data_target = wine_data.join(wine_target).set_index('target')
```

## ANALYSIS OF IRIS DATASET

```python
#PCA VISUALISATION
pc.principal_component_2d_timesample(data=iris_data_target,label='Target(iris class)')
#VISUALISATION OF EXPLAINED VARIANCE RATIO FOR PCs
pc.variance_ratio(iris_data_target)
#THE 'RECIPE' FOR FIRST PC IN FEATURE SPACE
pc.vectors_pc(iris_data_target, 1)
```

## ANALYSIS OF DIABETES DATASET

```python
#PCA VISUALISATION
pc.principal_component_2d_timesample(data=diabetes_data_target,label='Target(disease progression)', c_map='plasma', remove_legend=True)
#VISUALISATION OF EXPLAINED VARIANCE RATIO FOR PCs
pc.variance_ratio(diabetes_data_target)
#THE 'RECIPE' FOR PCs IN FEATURE SPACE
pc.vectors_pc(diabetes_data_target, 1)
pc.vectors_pc(diabetes_data_target, 2)
```

## ANALYSIS OF WINE DATASET

```python
#PCA VISUALISATION
pc.principal_component_3d_timesample(data=wine_data_target,label='Target(wine class)')
#VISUALISATION OF EXPLAINED VARIANCE RATIO FOR PCs
pc.variance_ratio(wine_data_target)
#THE 'RECIPE' FOR FIRST PC IN FEATURE SPACE
pc.vectors_pc(wine_data_target, 1)
```
